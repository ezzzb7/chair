```python
import pytz

# 创建一个新的数据框来存储转换后的时间（本地时间）
local_times_df = df.copy()

# 定义一个函数来转换时间到本地时区
def convert_to_local_time(row):
    try:
        tz_str = row['timezone']
        if tz_str == 'Unknown':
            return row['event_time']  # 对于未知时区，我们保留原始时间
        tz = pytz.timezone(tz_str)
        local_time = row['event_time'].astimezone(tz)
        return local_time.replace(tzinfo=None)  # 移除时区信息，只保留本地时间（如果需要的话）
    except Exception as e:
        # 处理可能的异常，例如时区字符串格式不正确
        # 在这里，我们选择返回原始时间，但也可以记录错误或进行其他处理
        return row['event_time']

# 应用函数转换时间
local_times_df['local_event_time'] = local_times_df.apply(convert_to_local_time, axis=1)

# 接下来，你可以按小时统计每个时区的活动情况
# 例如，统计每个时区每小时的活跃事件数（这里我们使用local_event_time的日期和时间部分）
hourly_activity = local_times_df.groupby([local_times_df['local_event_time'].dt.hour, local_times_df['timezone']]).size().unstack().fillna(0)

# 打印结果
print(hourly_activity)
```

# 2.协作行为分析

提交频率：统计每个用户的提交次数，识别高活跃用户和低活跃用户。

```python
import pandas as pd

# 读取CSV文件，使用原始文件
file_path = 'D:/users_combined_info_500.csv'
df = pd.read_csv(file_path)

# 统计每个用户的提交次数
user_submissions = df.groupby('user_id').size().reset_index(name='submission_count')

# 计算中位数作为阈值
threshold = user_submissions['submission_count'].median()

# 定义 bins，确保它们是唯一的
bins = [0, threshold, user_submissions['submission_count'].max()]
# 注意：由于 pd.cut() 是右开区间，max_value 将被包含在 [threshold, max_value) 区间内

# 识别高/低活跃用户
user_submissions['activity_level'] = pd.cut(user_submissions['submission_count'], 
                                            bins=bins, 
                                            labels=['Low Activity', 'High Activity'], 
                                            right=True)

# 打印结果
print(user_submissions)
```
